```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set(style="darkgrid")
```

```python
test_arr1 = [23, 16, 20, 24, 15, 11, 13, 2, 3, 22, 25, 12, 1, 19, 4, 5, 10, 18, 14, 8, 17, 21, 7, 6, 9]
num_threads = [1, 2, 4, 8, 16]
```

```python
len(test_arr1)
```

```python
test_arr2 = np.array(open('/home/dilara/CLionProjects/untitled5/IntegerArray.txt', 'r').read().split('\n'))
```

```python
len(test_arr2)
```

```python
'''SELECTION SORT'''
'''
23 16 20 24 15 11 13 2 3 22 25 12 1 19 4 5 10 18 14 8 17 21 7 6 9

Average time for 1 is: 0.000131629

Average time for 2 is: 0.000251609

Average time for 4 is: 0.00406696

Average time for 8 is: 0.00117173

Average time for 16 is: 0.00251553


'''

SelectionSort_res1 = np.array([0.000131629, 0.000251609, 0.00406696, 0.00117173, 0.00251553]) * 10/3


'''
Average time for 1 is: 4.98711

Average time for 2 is: 2.60098

Average time for 4 is: 2.95935

Average time for 8 is: 3.94173

Average time for 16 is: 5.50623

'''

SelectionSort_res2 = np.array([4.98711, 2.60098, 2.95935, 3.94173, 5.50623]) * 10 /3
```

```python
fig, axs = plt.subplots(figsize=(15, 7), ncols=2)
axs[0].plot(num_threads, SelectionSort_res1, c='g', label='test case 1 with n = %d' % len(test_arr1))
axs[0].scatter(num_threads, SelectionSort_res1, c='g')
axs[0].set_xlabel('Number of Threads')
axs[1].plot(num_threads, SelectionSort_res2, c='r', label='test case 2 with n = %d' % len(test_arr2))
axs[1].set_xlabel('Number of Threads')
axs[1].scatter(num_threads, SelectionSort_res2, c='r')
axs[0].legend()
axs[1].legend()
fig.suptitle('Selection Sort', fontsize=16)
plt.show()
```

```python
'''QUICK SORT'''
'''
23 16 20 24 15 11 13 2 3 22 25 12 1 19 4 5 10 18 14 8 17 21 7 6 9

 Average time for 1 is: 6.18013e-05

 Average time for 2 is: 0.000135296

 Average time for 4 is: 0.000486339

 Average time for 8 is: 0.000411374

 Average time for 16 is: 0.00109509
 
'''
QS_res1 = [6.18013e-05, 0.000135296, 0.000486339, 0.000411374, 0.00109509]

'''
test arr 2

 Average time for 1 is: 0.0564533

 Average time for 2 is: 0.117438

 Average time for 4 is: 0.0945461

 Average time for 8 is: 0.0680134

 Average time for 16 is: 0.0673668

'''

QS_res2 = [0.0564533, 0.117438, 0.0945461, 0.0680134, 0.0673668]
```

```python
fig, axs = plt.subplots(figsize=(15, 7), ncols=2)
axs[0].plot(num_threads, QS_res1, c='g', label='test case 1 with n = %d' % len(test_arr1))
axs[0].scatter(num_threads, QS_res1, c='g')
axs[0].set_xlabel('Number of Threads')
axs[1].plot(num_threads, QS_res2, c='r', label='test case 2 with n = %d' % len(test_arr2))
axs[1].set_xlabel('Number of Threads')
axs[1].scatter(num_threads, QS_res2, c='r')
axs[0].legend()
axs[1].legend()
fig.suptitle('Quick Sort', fontsize=16)
plt.show()
```

```python
'''MERGE SORT'''

'''
23 16 20 24 15 11 13 2 3 22 25 12 1 19 4 5 10 18 14 8 17 21 7 6 9

 Average time for 1 is: 0.000365095

 Average time for 2 is: 0.000328289

 Average time for 4 is: 0.000355652

 Average time for 8 is: 0.00078683

 Average time for 16 is: 0.00122572

'''

MS_res1 = [0.000365095, 0.000328289, 0.000355652, 0.00078683, 0.00122572]

'''
 Average time for 1 is: 0.366774

 Average time for 2 is: 0.266235

 Average time for 4 is: 0.251348

 Average time for 8 is: 0.206478

 Average time for 16 is: 0.213838
 
'''

MS_res2 = [0.366774, 0.266235, 0.251348, 0.206478, 0.213838]
```

```python
fig, axs = plt.subplots(figsize=(15, 7), ncols=2)
axs[0].plot(num_threads, MS_res1, c='g', label='test case 1 with n = %d' % len(test_arr1))
axs[0].scatter(num_threads, MS_res1, c='g')
axs[0].set_xlabel('Number of Threads')
axs[1].plot(num_threads, MS_res2, c='r', label='test case 2 with n = %d' % len(test_arr2))
axs[1].set_xlabel('Number of Threads')
axs[1].scatter(num_threads, MS_res2, c='r')
axs[0].legend()
axs[1].legend()
fig.suptitle('Merge Sort', fontsize=16)
plt.show()
```
